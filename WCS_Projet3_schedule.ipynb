```python
import requests
from requests.auth import HTTPBasicAuth
import schedule
import time
from datetime import datetime, timedelta
import pandas as pd
import requests, json
```

```python
def current_time():
    now = datetime.now()
    return now.strftime("_%H_%M_%S")
```

```python
current_time()
```

```python
def retrieve_data():
    
    username = 'estellebordevaire@yahoo.fr'
    password = 'projet3WCS!'

    r=requests.get('https://download.data.grandlyon.com/ws/rdata/tcl_sytral.tclpassagearret/all.json?maxfeatures=100&start=1',\
                   auth=HTTPBasicAuth(username, password))

    data=r.json()
    df=pd.json_normalize(data["values"])


    df.to_csv("passagearret" + current_time() + ".csv", index = False)

    return df


schedule.every(1).minutes.do(retrieve_data)

while True:
    schedule.run_pending()
    

# Ce script génère des fichiers "passagearret_XX_XX_XX.csv"
# Il doit être arrêté par CTRL + C
```
